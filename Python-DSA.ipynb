LINEAR-SEARCH

```python
def linearsearch(arr,target):
    n=len(arr)
    found=False
    for i in range(n):
        if (arr[i]==target):
            return i
            found=True
    if (found==False):
        return -1
array=[1,13,11,2,99,157,43,8,3]
x=3

print(linearsearch(array,x))
```

BINARY-SEARCH

```python
def binarysearch(arr,target):
    n=len(arr)
    start=0
    end=n
    if(arr[end-1]<target):
        return 'target is out of reach in max'
    if(arr[start]>target):
        return 'target is out of reach in min'
    while(start<end):
        mid=int((start+end)/2)
        if(arr[mid]==target):
            return mid
            end=-1
        elif(arr[mid]<target):
            start=mid
        else:
            end=mid
           
                
        
array=[1,3,5,7,9,11,22,33,44,55,66,777,8888,99999]    
x=-1
print(binarysearch(array,x))
        
```

Bubble sort

```python
def bubblesort(arr):
    n=len(arr)
    for i in range(0,n):
        for j in range(0,n-1):
            if(arr[j]>arr[j+1]):
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    return arr

testarray1=[9,8,7,6,5,4,3,12,10,2,1,11]
print("the sorted array is " , bubblesort(testarray1))
          
```

```python
def selectionsort(arr):
    n=len(arr)
    for i in range(0,n):
        minidx=i
        for j in range(i,n):
            if(arr[j]<arr[minidx]):
                minidx=j
        temp=arr[i]
        arr[i]=arr[minidx]
        arr[minidx]=temp
    return arr
myarr=[8,1,7,2,6,3,5,4]
print("the sorted array is : ",selectionsort(myarr))
```

```python
def factorial(num):
    if(num!=1):
        return num*factorial(num-1)
    else:
        return 1
print(factorial(10))
```

```python
def insertionsort(arr):
    for i in range(1,len(arr)):
        j=i-1
        key=arr[i]
        while(j>=0 and arr[j]>key):
            arr[j+1]=arr[j]
            j=j-1
        arr[j+1]=key
    return arr
myarr=[8,1,7,10,17,13,22,2,6,3,5,4]
print("the sorted array is : ",insertionsort(myarr))
            
```

```python
def quicksort(arr,low,high):
    if(low<high):
        pi=partition(arr,low,high)
        quicksort(arr,low,pi-1)
        quicksort(arr,pi+1,high)
    return arr
def partition(arr,low,high):
    pivot=arr[high]
    i=low-1
    for j in range (low,high):
        if(arr[j]<pivot):
            i=i+1
            temp=arr[i]
            arr[i]=arr[j]
            arr[j]=temp
    temp=arr[i+1]
    arr[i+1]=arr[high]
    arr[high]=temp
    return i+1
myarr=[8,6,1,5,90,12,4]
print("The sorted array is ", quicksort(myarr,0,len(myarr)-1))      
```

THE MERGESORT

```python
def mergesort(arr,left,right):
    if(left<right):
        mid=int((left+right)/2)
        mergesort(arr,left,mid)
        mergesort(arr,mid+1,right)
        merge(arr,mid,left,right)
    return arr
def merge(arr,middle,l,r):
    larr=arr[l:middle+1]
    rarr=arr[middle+1:r+1]
    i,j,k=0,0,l
    n1,n2=len(larr),len(rarr)
    while(i<n1 and j<n2):
        if(larr[i]<=rarr[j]):
            arr[k]=larr[i]
            i,k=i+1,k+1
        else:
            arr[k]=rarr[j]
            k,j=k+1,j+1
    while(i<n1):
        arr[k]=larr[i]
        k=k+1
        i=i+1
    while(j<n2):
        arr[k]=rarr[j]
        j=j+1
        k=k+1
    return arr
    
myarr=[6,1,10,4,16,23,2,5,32,11,22]
print("The unsorted array : ",myarr)
myarr=mergesort(myarr,0,len(myarr)-1)
print("the sorted array :",myarr)
```

SINGLE LINKEDLIST

```python
class node():
    def __init__(self,value,next):
        self.value=value
        self.next=next        
class linkedlist():
    def __init__(self):
        self.head = node(None,None)
    def add(self,data):
        if(self.head.value==None):
            self.head.value = data
        else:
            newnode = node(data,None)
            current = self.head
            while(current.next!=None):
                current=current.next
            current.next=newnode
    def returnall(self):
        current = self.head
        while(current!=None):
            print(current.value,end=' ')
            current=current.next
        print()
    def addfirst(self,value):
        newnode=node(value,self.head)
        self.head=newnode
    def delete(self,data):
        current=self.head
        while(current.next!=None and current.next.value!=data):
            current=current.next
        if(current.next!=None):
            current.next=current.next.next
mylist=linkedlist()
mylist.add(10)
mylist.add(11)
mylist.add(15)
mylist.returnall()
print("after inserted in front")
mylist.addfirst(7)
mylist.addfirst(5)
mylist.returnall()
print("after deletion of 11 ")
mylist.delete(11)
mylist.returnall()
```

DOUBLE LINKED LIST

```python
class node():
    def __init__(self,data,next,previous):
        self.data=data
        self.next=next
        self.previous=previous
class doublelinkedlist():
    def __init__(self):
        self.head=node(None,None,None)
        self.tail=None
    def add(self,value):
        if(self.head.data==None):
            self.head.data=value
            self.tail=self.head
        else:
            current=self.head
            newnode=node(value,None,None)
            while(current.next!=None):
                current=current.next
            current.next=newnode
            newnode.previous=current
            self.tail=newnode
    def printfromstart(self):
        current=self.head
        while(current!=None):
            print(current.data,end=' ')
            current=current.next
        print()
    def printfromend(self):
        current=self.tail
        while(current!=None):
            print(current.data,end=' ')
            current=current.previous
        print()
    def delete(self,value):
        current=self.head
        while(current.next!=None and current.next.data!=value):
            current=current.next
        if(current.next!=None):
            current.next.next.previous=current
            current.next=current.next.next
llst=doublelinkedlist()
print("double linked list")
llst.add(10)
llst.add(20)
llst.add(30)
llst.add(40)
llst.printfromstart()
print("reverse of double linked list")
llst.printfromend()
print ("after deletion  ")
llst.delete(30)
llst.printfromstart()
llst.printfromend()
       
```

STACK

```python
class Stack():
    def __init__(self,size):
        self.arr=[None]*size
        self.top=-1
        self.size=size
    def push(self,data):
        if(self.top<self.size-1):
            self.top+=1
            self.arr[self.top]=data
        else:
            print("stack overflowing...")
    def pop(self):
        if(self.top!=-1):
            self.top-=1
        else:
            print("stack underflowing...")
    def peek(self):
        if(self.top!=-1):
            print(self.arr[self.top])
        else:
            print("stack is empty...")
    def viewall(self):
        if(self.top!=-1):
            for i in range(0,self.top+1):
                print(self.arr[i],end=" ")
            print()
        else:
            print("stack is empty...")

bucket=Stack(5)
bucket.push(11)
bucket.push(22)
bucket.push(33)
bucket.push(44)
bucket.push(55)
bucket.push(66)
bucket.peek()
bucket.viewall()
bucket.pop()
bucket.viewall()
                       
```
